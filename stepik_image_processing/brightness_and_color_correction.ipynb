### Automatic contrast of colored image

task: https://stepik.org/lesson/60609/step/2

```python
task_img = img_as_float(imread("https://stepik.org/media/attachments/lesson/60609/tiger-color.png"))
y = ((0.2126, 0.7152, 0.0722) * task_img).sum(axis=2)
u = ((-0.0999, -0.3360, 0.4360) * task_img).sum(axis=2)
v = ((0.6150, -0.5586, -0.0563) * task_img).sum(axis=2)
ymin, ymax = np.percentile(y, q=(5, 95))
y = np.clip((y - ymin) / (ymax - ymin), 0.0, 1.0)
rgb = np.clip(np.dstack((y + 1.2803 * v, y - 0.2148 * u - 0.3805 * v, y + 2.1279 * u)), 0.0, 1.0)
imsave("out_img.png", img_as_ubyte(rgb))
```

### Gray world transformation

task: https://stepik.org/lesson/60610/step/2

```python
task_img = img_as_float(imread("https://stepik.org/media/attachments/lesson/60610/railroad.png"))
means = task_img.mean(axis=(0, 1))
avg = means.mean()
task_img = np.clip(task_img / (means / avg), 0.0, 1.0)
imsave("out_img.png", img_as_ubyte(task_img))
```

### Histogram equalization

task: https://stepik.org/lesson/60611/step/2

```python
task_img = imread("https://stepik.org/media/attachments/lesson/60611/landscape.png")
hist, bin_edges = np.histogram(task_img, bins=range(257))
cdf = hist.cumsum()
cdf_min = cdf[cdf.nonzero()].min()
timg = (255 * (cdf[task_img] - cdf_min) / (task_img.size - 1))
timg = np.round(timg).astype("uint8")
```
