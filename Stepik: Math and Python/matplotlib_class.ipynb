```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = range(-10, 10)
y = [element**2 for element in x]
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, 'r')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, 'r')
axes2 = fig.add_axes([0.3, 0.5, 0.4, 0.4])
axes2.plot(y, x, 'g')
fig.show()
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots(2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'g')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 8))
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'g')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 8))
axes[0][0].plot(x, y, 'r')
axes[0][1].plot(y, x, 'g')
axes[1][0].plot(x, x, 'r')
axes[1][1].plot(y, y, 'g')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.arange(0.1, 5, 0.1)


def f(x):
    return x**2

def g(x):
    return 1 / x

def h(x):
    return x * np.sin(x)
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), x, g(x))
axes.plot(x, h(x))
axes.legend(['f(x)', 'g(x)', 'h(x)'])
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), label=' f(x)')
axes.plot(x, g(x))
axes.plot(x, h(x), label=' h(x)')
axes.legend(loc=5)
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), label='$x^2$')
axes.plot(x, g(x), label=r'$\frac{1}{x}$')
axes.plot(x, h(x), label=r'$x \cdot sin(x)$')
axes.set_xlabel('x', fontsize=18)
axes.set_title('Title', fontsize=30)
axes.legend(loc=5, fontsize=18)
```
